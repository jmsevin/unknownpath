```python
import pandas as pd
```

# 1. Build dataset for Actors dashboard

```python
df_26_en = pd.read_csv("data/raw/TOPICS_LLM/COP26EN_FILTERED_combined_categorization.csv", dtype={'id': str})
df_26_en['cop'] = 26
df_26_en['lang'] = 'en'

df_26_fr = pd.read_csv("data/raw/TOPICS_LLM/COP26FR_EVALUATED.csv", dtype={'id': str})
df_26_fr['cop'] = 26
df_26_fr['lang'] = 'fr'

df_27_en = pd.read_csv("data/raw/TOPICS_LLM/COP27EN_FILTERED_combined_categorization.csv", dtype={'id': str})
df_27_en['cop'] = 27
df_27_en['lang'] = 'en'

df_27_fr = pd.read_csv("data/raw/TOPICS_LLM/COP27FR_EVALUATED.csv", dtype={'id': str})
df_27_fr['cop'] = 27
df_27_fr['lang'] = 'fr'

df_28_en = pd.read_csv("data/raw/TOPICS_LLM/COP28EN_FILTERED_combined_categorization.csv", dtype={'id': str})
df_28_en['cop'] = 28
df_28_en['lang'] = 'en'

df_28_fr = pd.read_csv("data/raw/TOPICS_LLM/COP28FR_EVALUATED.csv", dtype={'id': str})
df_28_fr['cop'] = 28
df_28_fr['lang'] = 'fr'

df_en = pd.concat([df_26_en, df_27_en, df_28_en], ignore_index=True)
df_fr = pd.concat([df_26_fr, df_27_fr, df_28_fr], ignore_index=True)
```

```python
df_en['categories'] = df_en['category1'] + '|' + df_en['category2']
df_fr['categories'] = df_fr['category1'] + '|' + df_fr['category2']
df_en['categories'] = df_en['categories'].str.replace('ERROR', '').str.replace('SKIPPED', '').str.split('|')
df_fr['categories'] = df_fr['categories'].str.split('|')
```

```python
df_en.dropna(subset=['categories'], inplace=True, ignore_index=True)
df_en = df_en.explode('categories', ignore_index=True,)
df_fr.dropna(subset=['categories'], inplace=True, ignore_index=True)
df_fr = df_fr.explode('categories', ignore_index=True)
```

```python
cats = ["Crise climatique", "Théories scientifiques", "Caractérisation de l’événement COP", "Communautés scientifiques", "Protestation", "Mesures d'urgence","Données scientifiques", "Finance", "Polarisation", "Technologies", "Climate Crisis", "Scientific Theories", "Characterization of the COP Event", "Scientific Communities", "Protest", "Emergency Measures","Scientific Data", "Polarization"]
df_en = df_en[df_en['categories'].isin(cats)].reset_index(drop=True)
df_fr = df_fr[df_fr['categories'].isin(cats)].reset_index(drop=True)
```

```python
cat_fr_to_cat_en = { "Crise climatique": "Climate Crisis", "Théories scientifiques": "Scientific Theories", "Caractérisation de l’événement COP": "Characterization of the COP Event", "Communautés scientifiques": "Scientific Communities", "Protestation": "Protest","Mesures d'urgence": "Emergency Measures", "Données scientifiques": "Scientific Data", "Finance": "Finance", "Polarisation": "Polarization", "Technologies": "Technologies" }
df_fr['categories'] = df_fr['categories'].map(cat_fr_to_cat_en)
```

```python
df = pd.concat([df_en, df_fr], ignore_index=True)
df.to_csv("data/processed/combined_categorization.csv", index=False, sep=';')
```
